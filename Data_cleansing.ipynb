```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from pandas import read_csv

data_ = pd.read_csv('F:/Research/Data/sms-call-internet-mi-2013-11-1.txt', delimiter='\t')

df = data_
df.head()

df.columns = ['Grid ID', 'Time Stamp','Caller ID','SMS in', 'SMSin','Call in','Call out','Internet']

# Fill all the NaN values with 0
df.fillna(0,inplace = True)

#The time is in milli-seconds. We will change it to minutes.
df['Time Stamp'] = df['Time Stamp']/(1000*60)

df.head()

plt.plot(df['Internet'])
plt.grid()
plt.xlabel('Time (in 10-minute interval)', fontsize =12)
plt.ylabel('Activity', fontsize = 12)
```
